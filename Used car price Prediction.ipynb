## Predicting the price of used car with given variables like Car Model, Milage and Age of the car.

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_excel('car dataset.xlsx')
data
```

#### Creading dummies using pandas to conver the categorical to numerical (binary values)

```python
dummies= pd.get_dummies(data['Car Model'])
dummies.head()
```

```python
merge= pd.concat((dummies,data),axis='columns')
merge=merge.drop((['Car Model']),axis='columns')
merge
```

```python
y=merge.drop(['Audi A5','BMW X5','Mercedez Benz C class','Mileage','Age(yrs)'],axis='columns')
y
```

#### Dropping one dummi variable to avoid multicollinearity issue.

```python
X=merge.drop(['Audi A5','Sell Price($)'],axis='columns')
X
```

```python
model= LinearRegression()
model.fit(X,y)
```

### Price prediction of Mercedez Benz C class 4 yr old with mileage 45000

```python
model.predict([[0,1,45000,4]])
```

### Price prediction of BMW X5 that is 7 yr old with mileage 86000

```python
model.predict([[1,0,86000,7]])
```

```python
print("prediction percentage: ", round((model.score(X,y)*100),2),"%")
```
